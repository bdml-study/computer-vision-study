# Read / Write Image using OpenCV

```python
import cv2
import matplotlib.pylab as plt
```

```python
image_gray = cv2.imread("../data/lena.jpeg", cv2.IMREAD_GRAYSCALE)
image_gray.shape
```

```python
plt.imshow(image_gray, cmap="gray")
plt.show()
```

```python
image_color = cv2.imread("../data/lena.jpeg", cv2.IMREAD_COLOR)
print(image_color.shape) # 무조건 3채널 (alpha 있으면 날려버림)

image_colora = cv2.imread("../data/lena.jpeg", cv2.IMREAD_UNCHANGED)
print(image_colora.shape) # alpha가 있으면 4채널로 표시됨
```

```python
plt.imshow(image_color)
plt.show()
```

### Convert BGR 2 RGB

```python
image_color = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)
plt.imshow(image_color)
plt.show()
```

### Writing

```python
cv2.imwrite("./temp.jpg", image_gray)
```
